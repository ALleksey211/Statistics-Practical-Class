# Basics of Python
Python is a high-level, dynamically typed multiparadigm programming language. Python code is often said to be almost like pseudocode, since it allows you to express very powerful ideas in very few lines of code while being very readable. As an example, here is an implementation of the classic quicksort algorithm in Python:

````python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

print(quicksort([3,6,8,10,1,2,1]))
````

#### Slicing
In addition to accessing list elements one at a time, Python provides concise syntax to access sublists; this is known as slicing:

````python
nums = list(range(5))    # range is a built-in function that creates a list of integers
print(nums)         # Prints "[0, 1, 2, 3, 4]"
print(nums[2:4])    # Get a slice from index 2 to 4 (exclusive); prints "[2, 3]"
print(nums[2:])     # Get a slice from index 2 to the end; prints "[2, 3, 4]"
print(nums[:2])     # Get a slice from the start to index 2 (exclusive); prints "[0, 1]"
print(nums[:])      # Get a slice of the whole list; prints ["0, 1, 2, 3, 4]"
print(nums[:-1])    # Slice indices can be negative; prints ["0, 1, 2, 3]"
nums[2:4] = [8, 9] # Assign a new sublist to a slice
print(nums)         # Prints "[0, 1, 8, 9, 4]"
````

#### List comprehensions

When programming, frequently we want to transform one type of data into another. As a simple example, consider the following code that computes square numbers:

````python
nums = [0, 1, 2, 3, 4]
squares = []
for x in nums:
    squares.append(x ** 2)
print(squares)
````

You can make this code simpler using a list comprehension:

````python
nums = [0, 1, 2, 3, 4]
squares = [x ** 2 for x in nums]
print(squares)
````

List comprehensions can also contain conditions:

````python
nums = [0, 1, 2, 3, 4]
even_squares = [x ** 2 for x in nums if x % 2 == 0]
print(even_squares)
````

## Functions

Python functions are defined using the `def` keyword. For example:

````python
def sign(x):
    if x > 0:
        return 'positive'
    elif x < 0:
        return 'negative'
    else:
        return 'zero'

for x in [-1, 0, 1]:
    print(sign(x))
````

### Task: Course Management System for Students

#### Description

Develop a system to manage course enrollments for students at a university. This system should allow students to enroll in courses, drop courses, and list currently enrolled courses. Additionally, the system should enable the management of course information and track the history of student enrollments and course completions.

#### Tasks

1. **`Course` Class**:
   - Attributes: course name, course code, maximum students, list of enrolled students.
   - Methods: constructor, method to add a student to the course, method to remove a student from the course, method to display course info.

2. **`Student` Class**:
   - Attributes: student name, student ID, list of enrolled courses.
   - Methods: constructor, methods for enrolling in a course, dropping a course, listing currently enrolled courses.

3. **`EnrollmentSystem` Class**:
   - Attributes: list of courses, list of students, enrollment history (which student enrolled in/dropped which course and when).
   - Methods: constructor, methods for adding and removing courses, registering and deregistering students, methods for enrolling and dropping courses for students, method to display enrollment history.

#### Assignment

1. Implement the `Course`, `Student`, and `EnrollmentSystem` classes with the specified attributes and methods.
2. Create several courses and students, and add them to the enrollment system.
3. Implement scenarios for students enrolling in and dropping courses.
4. Display the enrollment history to show the sequence of enrollments and course completions.

This task focuses on object-oriented programming principles, encouraging the practice of managing relationships between objects (courses and students) within a larger system (the enrollment system).


### Solution

````python
class Course:
    def __init__(self, name, code, max_students):
        self.name = name
        self.code = code
        self.max_students = max_students
        self.enrolled_students = []
        print(f"Course {name} created")

    def add_student(self, student):
        if len(self.enrolled_students) < self.max_students and student not in self.enrolled_students:
            self.enrolled_students.append(student)
            print(f"{student.name} enrolled in {self.name}")
            return True
        print(f"Enrollment failed. Student limit reached for {self.name}")
        return False

    def remove_student(self, student):
        if student in self.enrolled_students:
            self.enrolled_students.remove(student)
            print(f"{student.name} dropped {self.name}")
            return True
        print(f"Drop failed. {student.name} is not enrolled in {self.name}")
        return False

    def display_info(self):
        return f"{self.name} ({self.code}) - Max Students: {self.max_students}, Enrolled: {len(self.enrolled_students)}"

class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.enrolled_courses = []
        print(f"Student {name} created")

    def enroll(self, course):
        if course.add_student(self):
            self.enrolled_courses.append(course)
            return True
        return False

    def drop(self, course):
        if course.remove_student(self):
            self.enrolled_courses.remove(course)
            return True
        return False

    def list_courses(self):
        return [course.display_info() for course in self.enrolled_courses]

class EnrollmentSystem:
    def __init__(self):
        self.courses = []
        self.students = []
        self.enrollment_history = []
        print("Enrollment system created")

    def add_course(self, course):
        self.courses.append(course)

    def remove_course(self, course):
        if course in self.courses:
            self.courses.remove(course)

    def register_student(self, student):
        self.students.append(student)

    def deregister_student(self, student):
        if student in self.students:
            self.students.remove(student)

    def enroll_student(self, student, course):
        if student.enroll(course):
            self.enrollment_history.append(f"{student.name} enrolled in {course.name}")
            return True
        return False

    def drop_student(self, student, course):
        if student.drop(course):
            self.enrollment_history.append(f"{student.name} dropped {course.name}")
            return True
        return False

    def display_enrollment_history(self):
        for record in self.enrollment_history:
            print(record)

# Example usage
system = EnrollmentSystem()

# Creating courses
course1 = Course("Introduction to Python", "CS101", 30)
system.add_course(course1)

# Registering students
student1 = Student("Alice Johnson", "S1001")
system.register_student(student1)

# Enrolling in courses
system.enroll_student(student1, course1)

# Dropping courses
system.drop_student(student1, course1)

# Displaying enrollment history
system.display_enrollment_history()
````

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
def rotate_matrix(matrix, text):
    if(text=='right'):
        return[col[::-1] for col in zip(*matrix)]
    if(text=='left'):
        return[col for col in zip(*matrix)][::-1]
    else:
        return(-1)

matrix=[[1,2,3],
        [4,5,6],
        [7,8,9]]
print('Поворот направо')
for row in rotate_matrix(matrix,'right'):
    print(row)

print('Поворот налево')
for row in rotate_matrix(matrix,'left'):
    print(row)
````

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
import datetime
from typing import List, Optional

class Book: # Книги
    def __init__(self, title: str, author: str, year: int, isbn: str, copies: int):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.copies = copies  # количество доступных копий

    def get_info(self) -> str:
        return f"'{self.title}' {self.author} ({self.year}), ISBN: {self.isbn}, доступно копий: {self.copies}"

    def change_copies(self, delta: int) -> bool:
        new_copies = self.copies + delta
        if new_copies < 0:
            print(f"Ошибка: нельзя выдать больше копий, чем есть (в наличии {self.copies})")
            return False
        self.copies = new_copies
        return True


class User: # Полььзователи

    def __init__(self, name: str, library_card_number: str):
        self.name = name
        self.library_card_number = library_card_number
        self.borrowed_books: List[Book] = []  # список взятых книг
        self.register()  # метод регистрации

    def register(self) -> None:
        print(f"Пользователь '{self.name}' (карта №{self.library_card_number}) зарегистрирован.")

    def add_borrowed_book(self, book: Book) -> None: # Взятие книги
        self.borrowed_books.append(book)

    def remove_borrowed_book(self, book: Book) -> bool: # Возврат кнмиги
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            return True
        return False

    def get_borrowed_books(self) -> List[str]: # Список взятых книжек
        return [book.title for book in self.borrowed_books]


class Library: # Библтотека

    def __init__(self):
        self.books: List[Book] = [] # все книги в библиотеке
        self.users: List[User] = []  # все зарегистрированные пользователи
        self.transaction_history: List[str] = []  # история операций

    def add_book(self, book: Book) -> None: # добавление книги в библиотеку
        self.books.append(book)
        print(f"Книга добавлена: {book.get_info()}")

    def delete_book(self, isbn: str) -> bool: # Удаление книги(по ibsn)
        for book in self.books:
            if book.isbn == isbn:
                self.books.remove(book)
                print(f"Книга '{book.title} {book.ibsn}' удалена из библиотеки.")
                return True
        print(f"Книга с ISBN {isbn} не найдена.")
        return False

    def add_user(self, user: User) -> None: # регистрация пользователя
        self.users.append(user)
        print(f"Пользователь '{user.name}' добавлен в систему.")

    def delete_user(self, library_card_number: str) -> bool: # удаление полльзователя
        for user in self.users:
            if user.library_card_number == library_card_number:
                self.users.remove(user)
                print(f"Пользователь '{user.name} {user.library_card_number}' удалён из системы.")
                return True
        print(f"Пользователь с картой №{library_card_number} не найден.")
        return False

    def issue_book(self, isbn: str, library_card_number: str) -> bool: # выдача книги пользователю
        # Поиск книги
        book = None
        for b in self.books:
            if b.isbn == isbn:
                book = b
                break
        if not book:
            print(f"Книга с ISBN {isbn} не найдена в библиотеке.")
            return False

        # Поиск пользователя
        user = None
        for u in self.users:
            if u.library_card_number == library_card_number:
                user = u
                break
        if not user:
            print(f"Пользователь с картой №{library_card_number} не зарегистрирован.")
            return False

        if book.copies <= 0:
            print(f"Нет доступных копий книги '{book.title}'.")
            return False

        # Выдача книги
        if book.change_copies(-1):  # уменьшаем количество копий
            user.add_borrowed_book(book)
            history_entry = f"[ВЫДАЧА]: Пользователь '{user.name}' взял книгу '{book.title}' (ISBN: {book.isbn})"
            self.transaction_history.append(history_entry)
            #print(history_entry)
            return True
        return False

    def return_book(self, isbn: str, library_card_number: str) -> bool: # возврат книги
        # Поиск книги
        book = None
        for b in self.books:
            if b.isbn == isbn:
                book = b
                break
        if not book:
            print(f"Книга с ISBN {isbn} не найдена в библиотеке.")
            return False

        # Поиск пользователя
        user = None
        for u in self.users:
            if u.library_card_number == library_card_number:
                user = u
                break
        if not user:
            print(f"Пользователь с картой №{library_card_number} не зарегистрирован.")
            return False


        if book not in user.borrowed_books:
            print(f"Пользователь '{user.name}' не брал книгу '{book.title}'.")
            return False

        # Возврат книги
        book.change_copies(1)
        user.remove_borrowed_book(book)
        history_entry = f"[ВОЗВРАТ]: Пользователь '{user.name}' вернул книгу '{book.title}' (ISBN: {book.isbn})"
        self.transaction_history.append(history_entry)
        print(history_entry)
        return True


    def search_books(self, title: Optional[str] = None, author: Optional[str] = None, year: Optional[int] = None, isbn: Optional[str] = None) -> List[Book]: # Поиск книги по любому из параметров
        results = self.books
        if title:
            results = [b for b in results if title.lower() in b.title.lower()]
        if author:
            results = [b for b in results if author.lower() in b.author.lower()]
        if year:
            results = [b for b in results if b.year == year]
        if isbn:
            results = [b for b in results if b.isbn == isbn]
        return results

    def search_users(self, name: Optional[str] = None, card_number: Optional[str] = None) -> List[User]: # Поиск пользователя по имени/номеру карты
        results = self.users
        if name:
            results = [u for u in results if name.lower() in u.name.lower()]
        if card_number:
            results = [u for u in results if u.library_card_number == card_number]
        return results

    def display_transaction_history(self) -> None: # История
        if not self.transaction_history:
            print("История операций пуста.")
            return
        print("\nИСТОРИЯ ОПЕРАЦИЙ БИБЛИОТЕКИ: ")
        for entry in self.transaction_history:
            print(entry)
        print("-"*100 + "\n")


# ДЕМОНСТРАЦИЯ РАБОТЫ СИСТЕМЫ
if __name__ == "__main__":
    # Создаём библиотеку
    library = Library()

    # Создаём несколько книг
    book1 = Book("Война и мир", "Лев Толстой", 1869, "978-5-17-118889-7", 3)
    book2 = Book("Преступление и наказание", "Фёдор Достоевский", 1866, "978-5-17-118891-0", 2)
    book3 = Book("Мастер и Маргарита", "Михаил Булгаков", 1967, "978-5-17-118893-4", 1)
    book4 = Book("1984", "Джордж Оруэлл", 1949, "978-5-17-118894-1", 0)  # нет копий

    # Добавляем книги в библиотеку
    library.add_book(book1)
    library.add_book(book2)
    library.add_book(book3)
    library.add_book(book4)

    # Создаём пользователей
    user1 = User("Иван Петров", "LIB-001")
    user2 = User("Мария Сидорова", "LIB-002")
    user3 = User("Петя Пупкин", "LIB-003")

    # Добавляем пользователей в библиотеку
    library.add_user(user1)
    library.add_user(user2)
    library.add_user(user3)

    # Поиск книг по автору
    print("\nПоиск книг Достоевского:")
    dostoevsky_books = library.search_books(author="Достоевский")
    for b in dostoevsky_books:
        print(f"  - {b.get_info()}")

    # Поиск пользователей по имени
    print("\nПоиск пользователей с именем 'Иван':")
    found_users = library.search_users(name="Иван")
    for u in found_users:
        print(f"  - {u.name}, карта №{u.library_card_number}")

    # Сценарий выдачи и возврата книг
    print("\n--- Операции выдачи ---")
    # Выдача книги, которая есть в наличии
    library.issue_book("978-5-17-118889-7", "LIB-001")
    library.issue_book("978-5-17-118891-0", "LIB-001")
    library.issue_book("978-5-17-118893-4", "LIB-002")
    # Попытка выдать книгу с нулевым количеством копий
    library.issue_book("978-5-17-118894-1", "LIB-003")
    # Попытка выдать несуществующую книгу
    library.issue_book("999-9-99-999999-9", "LIB-001")

    print("\n--- Операции возврата ---")
    # Возврат книги
    library.return_book("978-5-17-118891-0", "LIB-001")
    library.return_book("978-5-17-118893-4", "LIB-002")
    # Попытка вернуть книгу, которую пользователь не брал
    library.return_book("978-5-17-118893-4", "LIB-003")

    # Отображение итоговой истории операций
    library.display_transaction_history()

    # Вывод информации о текущем состоянии книг и пользователей
    print("Текущее состояние книг:")
    for book in library.books:
        print(f"  {book.get_info()}")

    print("\nТекущие долги пользователей:")
    for user in library.users:
        borrowed = user.get_borrowed_books()
        if borrowed:
            print(f"  {user.name}: {', '.join(borrowed)}")
        else:
            print(f"  {user.name}: нет книг на руках")
````
